# Task 1: Exploring and Visualizing Iris Dataset

## Problem Statement
The objective of this task is to explore and visualize the Iris dataset to understand patterns, relationships, and distributions of features using data analysis techniques.

## Dataset Overview
The dataset contains 150 samples with 5 columns including four numerical features (sepal_length, sepal_width, petal_length, petal_width) and one categorical variable (species). There are no missing values in the dataset.

## Data Visualization Explanation
- Scatter plot shows the relationship between sepal length and petal length for different species.
- Histograms show the distribution of each feature.
- Box plot helps identify outliers in the dataset.

## Conclusion
The Iris dataset shows clear distinctions between species based on petal and sepal measurements. Data visualization helps in understanding feature relationships, distributions, and identifying outliers effectively.

```python
# ===============================
# Task 1: Exploring Iris Dataset
# ===============================

# Import Libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load Dataset
df = sns.load_dataset('iris')

# Show basic info
print("Shape of dataset:", df.shape)

print("\nColumn Names:")
print(df.columns)

print("\nFirst 5 rows:")
print(df.head())

# Dataset summary
print("\nDataset Info:")
df.info()

print("\nStatistical Summary:")
print(df.describe())

# Scatter Plot
plt.figure()
sns.scatterplot(x='sepal_length', y='petal_length', hue='species', data=df)
plt.title("Scatter Plot of Sepal Length vs Petal Length")
plt.show()

# Histograms
df.hist(figsize=(10,8))
plt.suptitle("Histogram of Features")
plt.show()

# Box Plot
plt.figure()
sns.boxplot(data=df)
plt.title("Box Plot for Outlier Detection")
plt.show()

# Conclusion
print("\nConclusion:")
print("The dataset shows clear differences between iris species. Visualization helps understand feature relationships and detect outliers.")
```
